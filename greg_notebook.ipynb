```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
water_features = pd.read_csv("https://data.wprdc.org/datastore/dump/513290a6-2bac-4e41-8029-354cbda6a7b7", index_col = "_id")
df = water_features[["neighborhood","feature_type"]].sort_values(by="neighborhood")
df_vals = df["neighborhood"].value_counts() 
df_vals.plot(kind="bar", figsize=(20,8), xlabel="Neighborhood", ylabel="Number of Water Features")
```
