```python
#Importing nessecary Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing and assigning file containing median age of death per neighboorhood
df = pd.read_csv("https://data.wprdc.org/dataset/ee735209-4de7-4ea4-b446-bf0f0f6d9cb3/resource/c2e1500a-a12a-4e91-be94-76c6a892b7e2/download/nhoodmedianage20112015.csv")
```

```python
#Creating a series and defining what all information should be printed
median_age_series = (
    df.groupby("NEIGHBORHOOD")["TOTAL MD AGE AT DEATH"]
    .median()
    .fillna(0)
)

#Displaying data on every neighboorhood
pd.set_option("display.max_rows", None)

#Sorting the data set from highest to lowest
ranked_median_age = median_age_series.sort_values(ascending=False)

#Displaying the series
ranked_median_age
```

```python
#Creating graph

#defining graph dimensions
plt.figure(figsize=(10, 12))

#Defining what kind of graphical representation should be made
median_age_series.sort_values().plot(kind="barh")

#Assigning titles to the graph
plt.title("Median Age of Death by Neighborhood in Pittsburgh")
plt.xlabel("Median Age of Death")
plt.ylabel("Neighborhood")

#Displaying the graph
plt.tight_layout()
plt.show()
```
